```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df1 = pd.read_csv("/home/emmo/Desktop/Data_Visual/house-prices-advanced-regression-techniques/test.csv")
df1.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df1.isnull())
```

```python
percentage_of_cols = (df1.isnull().sum() / df1.shape[0]) * 100
percentage_of_cols
```

```python
more_missing_val_cols = percentage_of_cols[percentage_of_cols > 15].keys()
more_missing_val_cols
```

```python
df2 = df1.drop(columns = more_missing_val_cols)
df2.shape
```

```python
plt.figure(figsize=(25, 25))
sns.heatmap(df2.isnull())
```

```python
df2_per = (df2.isnull().sum() / df2.shape[0])* 100
df2_per
```

```python
df3_numeric = df2.select_dtypes(include = ['int64', 'float64'])
df3_numeric
```

```python
df3_missing_cols = [var for var in df3_numeric.columns if df3_numeric[var].isnull().sum() > 0]
df3_missing_cols
```

```python
df4_mean = df3_numeric.fillna(df3_numeric.mean())
df4_mean.isnull().sum().sum()
```

```python
df5_medin = df3_numeric.fillna(df3_numeric.median())
```

```python
plt.figure(figsize=(25,16))
for i ,var in enumerate(df3_missing_cols):
    plt.subplot(4,3,i+1)
    sns.distplot(df3_numeric[var], hist = False , bins = 20)
    sns.distplot(df4_mean[var], hist = False,  bins = 20)
    sns.distplot(df5_medin[var], hist = False,  bins = 20)
  
```

```python
df6_concated = pd.concat([ df3_numeric [df3_missing_cols], 
          df4_mean[df3_missing_cols],
          df5_medin[df3_missing_cols]], axis = 1)
```

```python
df6_concated[df6_concated.isnull().any(axis = 1)]
```
