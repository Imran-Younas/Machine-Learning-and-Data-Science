```python
percentage_cols = (df1.isnull().sum() / df1.shape[0] ) * 100
percentage_cols
```

```python
more_missing_value_cols = percentage_cols[percentage_cols > 20].keys()
more_missing_value_cols
```

```python
df2 = df1.drop(columns = more_missing_value_cols)
df2.shape
```

```python
df3_numeric = df2.select_dtypes(include=['int64', 'float64'])
```

```python
missing_numeric_cols = [var for var in df3_numeric.columns if df3_numeric[var].isnull().sum() > 0]
missing_numeric_cols
```

```python

plt.figure(figsize=(25,25))
for i ,var in enumerate(missing_numeric_cols):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_numeric[var], bins = 20)
```

```python
df4_numeric_mean = df3_numeric.fillna(df3_numeric.mean())
df4_numeric_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(25,25))
for i ,var in enumerate(missing_numeric_cols):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_numeric[var], bins = 20)
    sns.distplot(df4_numeric_mean[var], bins = 20)
```

```python
df5_numeric_median = df3_numeric.fillna(df3_numeric.median())
df5_numeric_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(25,25))
for i ,var in enumerate(missing_numeric_cols):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_numeric[var], bins = 20, hist= False)
    sns.distplot(df4_numeric_mean[var], bins = 20, hist= False)
    sns.distplot(df5_numeric_median[var], bins = 20, hist= False)
```

```python
df_concat = pd.concat([df3_numeric[missing_numeric_cols],
           df4_numeric_mean[missing_numeric_cols],
           df5_numeric_median[missing_numeric_cols]], axis = 1)
```
