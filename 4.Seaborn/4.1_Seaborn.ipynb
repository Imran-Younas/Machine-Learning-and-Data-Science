## Barplot or BarGraph Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day,
           y = tips_df.total_bill,
           hue = tips_df.sex,         # hue perameter use for seprating colums
           )
```

```python
sns.barplot(x = 'day',
           y = 'total_bill',
           hue = 'sex',         # hue perameter use for seprating colums
           data = tips_df        # same as above just make it simple using data permeeter
           )
plt.title('Barplot', fontsize = 20)
plt.xlabel('Days'),
plt.ylabel('Total Bill')

plt.show()
    
```

## Seaborn scatter plot

```python
dataset = sns.load_dataset("titanic")
dataset
```

```python
sns.scatterplot(x = "age", y = "fare", data = dataset)
plt.show()
```

```python
sns.scatterplot(x = "age", 
                y = "fare", 
                data = dataset,
                hue = "sex",      # differeniate category
                style = "who",      # assign different markers
                size = "who",       # size on base of assinding order
                sizes = (50, 200)    # input size
                
               )
```

```python
# mix some plot to see amazing result just for fun 


sns.scatterplot(x = "age", 
                y = "fare", 
                data = dataset,
                hue = "sex",    
                style = "who",      
                size = "who",      
                sizes = (50, 200)    
                
               )

sns.barplot(x = "age", y = "fare", data = dataset, hue = "sex")
```

```python
globel_warming = pd.read_csv("/home/imran/Downloads/globel.csv")
globel_warming
```

```python
gw = globel_warming.drop(columns= ["Country Code","Indicator Name", "Indicator Code"], axis = 1).set_index("Country Name")
gw
```

```python
sns.heatmap(gw)
```

```python
plt.figure(figsize=(16,7))
sns.heatmap(gw, 
            cmap = "coolwarm"
            , annot= True     # show quantity of color in each box
           )
```

```python
plt.figure(figsize=(16,7))
sns.heatmap(gw, 
            cmap = "coolwarm",
             annot= True,    # show quantity of color in each box
           linewidth = 2,    # size of line between 2 boxes
           linecolor = 'k',   # color of line between boxes
           cbar = False,       # hide the color side bar 
            
           )
```
