## Seaborn Line plot

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats as norm
```

```python
days = [1,2,3,4,5,6,7]
temperature = [33,34,32,36,37,31,33]
df_data = pd.DataFrame({"days": days, "temperature": temperature})
sns.lineplot(x = "days", y = "temperature", data = df_data)    # seaborn are taking data from datasets not from single list
plt.show()
```

```python
tips_df = sns.load_dataset("tips")    # load_dataset mean import data from github
tips_df
```

```python
df1 = sns.lineplot(x = "total_bill", y = "tip", data = tips_df)
plt.show()
```

```python
df1 = sns.lineplot(x = "total_bill", y = "size", data = tips_df)
plt.show()
```

```python
df1 = sns.lineplot(x = "size", y = "total_bill", data = tips_df)
plt.show()
```

```python
df1 = sns.lineplot(x = "total_bill", y = "size", data = tips_df,
                  hue = "day",      #difference between category  and also show as legend in grph
                  style = "day",     # change style of each line
                  palette = "hot",    # color of each line
                  markers =  ["<", "v", "o", ">"],   # marker in list form
                  legend ='full'     # legend value u can off from here using False permeter
                   
          )

plt.title("Tips", fontsize = 20)
plt.xlabel("total_bill", fontsize = 15)
plt.ylabel("size", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=([16,9]))    #size of graph
sns.set(style='darkgrid')

df1 = sns.lineplot(x = "total_bill", y = "size", data = tips_df,
                  hue = "day",      #difference between category  and also show as legend in grph
                  style = "day",     # change style of each line
                  palette = "hot",    # color of each line
                  markers =  ["<", "v", "o", ">"],   # marker in list form
                  legend ='full'     # legend value u can off from here using False permeter
                   
          )

plt.title("Tips", fontsize = 20)
plt.xlabel("total_bill", fontsize = 15)
plt.ylabel("size", fontsize = 15)
plt.savefig("hello.png")   # save the figure
plt.show()
```

## Seaborn Histogram / Distplot

```python
df1 = sns.load_dataset("tips")
sns.displot(df1["size"])      # displot function using for histogramb bB
```

```python
df1 = sns.load_dataset("tips")
sns.distplot(df1["size"],      # displot function using for histogramb bB
           bins = 100,
           hist = False    # take boolean values to visable or hide the histogram
           )

plt.show()
```

```python
df1 = sns.load_dataset("tips")
sns.displot(df1["size"],      # displot function using for histogramb bB
           kde = False     # remove kde line. it also take boolean value
         
           )

plt.show()
```

```python
df1 = sns.load_dataset("tips")
sns.displot(df1["size"],      # displot function using for histogramb bB
             kde = True,     # remove kde line. it also take boolean value
             rug = True,     # rug permeter show us the exact loction of x-axis from where histogram bar is started. it also take boolean value
             
    
         
           )

plt.show()
```

```python
df1 = sns.load_dataset("tips")
sns.displot(df1["size"],      # displot function using for histogramb bB
             kde = True,     # remove kde line. it also take boolean value
             rug = True,     # rug permeter show us the exact loction of x-axis from where histogram bar is started. it also take boolean value
             color = "r"      # color of graph
    
         
           )

plt.show()
```
