## Histograme

```python
ml_age = np.random.randint(20, 60, (50))   # create 50 number random data
py_age = np.random.randint(10, 30, (50))   # create 50 number random data

print(ml_age)
print(py_age)
```

```python
bins = [10,20,30,40,50,60]
plt.hist(ml_age, bins, rwidth = 0.9, histtype = "step")  #histype perameter
plt.title("Ages histogram")
plt.xlabel("ages")
plt.ylabel("People")
plt.show()
```

```python
bins = [10,20,30,40,50,60]
plt.hist(ml_age, bins, rwidth = 0.9,   # rwidth perameter make space btw 2 bar graphs, bins perameter to make range of bar
         orientation = "horizontal",   # orientation horizontal
         color = "black",   # color of bars
         label = "Ml_student")  
plt.title("Ages histogram")
plt.xlabel("ages")
plt.ylabel("People")
plt.legend()
plt.show()
```

```python
# 2 bars graphs 
plt.figure(figsize = [17,12])   # size of histogram
bins = [10,20,30,40,50,60]
plt.hist([ml_age, py_age], bins, rwidth = 0.9,   # rwidth perameter make space btw 2 bar graphs, bins perameter to make range of bar
         orientation = "vertical",   # orientation horizontal
         color = ["black", "red"],   # color of bars
         label = ["Ml_student", "Python_Student"])  
plt.title("Ages histogram")
plt.xlabel("ages")
plt.ylabel("People")
plt.legend()
plt.show()
```

## Pie Chart

```python
class_course = np.array(["Cyber Security", "Image Processin",
                         "Machine Learning", "Web devolopment", "App development"])
class_student = np.array([45,50,30,65,70])
```

```python
plt.pie([1])  # just print chart only of single value
plt.show()
```

```python
exp = [0,0,0,0.1,0]   # give infomation to seprate the partion from chart, value must be equal number of data
clr = ["r", "m", "y", "c", "g"]
textprops = {"fontsize": 18}  # pass value of text size of text in chart


plt.pie(class_student,   # print pie chart of different values
      labels = class_course, # give label to all prtitions
       labeldistance = 1.2,   # labels distance from radius
       explode = exp,   # explode here
       colors = clr ,    # change colors of pie chart of each partition
       autopct = "%0.2f%%",   # give value in percentage form "0.2" mean two float digits
       pctdistance = 0.5,   # distance of percentage from radius
       shadow = True,   # shadow of chart
       radius = 1.5,   # size of chart
       startangle = 45,    # startangle mean. from which angle do you want to start pass that angle value to startangle perameter
       textprops = textprops,   # text size of chart
       counterclock = True,     # clock or anticlock wise

       )  
plt.show()
```

```python
plt.figure(figsize = (8,4))
exp = [0,0,0,0.1,0]   # give infomation to seprate the partion from chart, value must be equal number of data
clr = ["r", "m", "y", "c", "g"]
textprops = {"fontsize": 18}  # pass value of text size of text in chart
wedge = {"linewidth": 4, "width": 1, "edgecolor": "k"}

plt.pie(class_student,   # print pie chart of different values
      labels = class_course, # give label to all prtitions
       labeldistance = 1.2,   # labels distance from radius
       explode = exp,   # explode here
       colors = clr ,    # change colors of pie chart of each partition
       autopct = "%0.2f%%",   # give value in percentage form "0.2" mean two float digits
       pctdistance = 0.5,   # distance of percentage from radius
       shadow = True,   # shadow of chart
       radius = 1.5,   # size of chart
       startangle = 45,    # startangle mean. from which angle do you want to start pass that angle value to startangle perameter
       textprops = textprops,   # text size of chart
       counterclock = True,     # clock or anticlock wise
       wedgeprops = wedge, # leanth, width edges sizes 
       center = (1, 6) ,   # away from center e.g 1st x value and 2nd y value 
       frame = True,   # give frame to chart
       rotatelabels = True   # rotate the text 
       )   
plt.show()
```

## How to save plots and charts

```python
data = [1,2,3,4]
names = ["a", "b", "c", "d"]
plt.pie(data, 
       labels = names)
plt.savefig("fig", #name of file
           dpi = 100,  # dpi mean dot per inch use for resolution of plot
           quality = 50,  # quatlity of plot value will be between 1 - 100
           facecolor = 'y',    # background color of charts
           )
plt.show()
```
