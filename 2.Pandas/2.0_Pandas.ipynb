## Pandas fillna()

```python
df = pd.read_csv('/home/imran/Downloads/100 Sales Records.csv')
df
```

```python
df.fillna(0)   # fill NA value on any scalar value e.g 0,1,2,3
```

```python
df.fillna(1)   # fill NA value on any scalar value e.g 0,1,2,3
```

```python
df.fillna(19)   # fill NA value on any scalar value e.g 0,1,2,3
```

```python
df.fillna({'ID': 'none', 'profit': 0}) #  use dictonory  replace  multiple  columns as as string
```

```python
df.fillna(method = 'ffill')   # filled value same as before value columns wsie
```

```python
df.fillna(method = 'bfill') # filled value same as after value columns wsie
```

```python
df.fillna(method = 'bfill', axis = 1) # filled value same as after value row wsie
```

```python
df.fillna(method = 'ffill', axis = 1) # filled value same as before value row wsie
```

```python
df.fillna(2, limit = 1)   # give limit to code that fill limited cell of int e.g here fill only 1 cell in each column
```

```python
df.fillna(3, limit = 7)   # give limit to code that fill limited cell of int e.g here fill only 1 cell in each column
```

```python
df.fillna(axis = 1,method = 'ffill', limit = 7) # also use axis and method for limit
```

```python
df.fillna(5 , inplace = True)  # inplace values as u wish
df 
```

## Pandas replace()

```python
df = pd.read_csv('/home/imran/Downloads/100 Sales Records.csv')
df
```

```python
df.replace(to_replace = 'hi', value = 'Hello') # replace values
```

```python
df.replace('Tuvalu', 'Never') # also use this method for replace 
```

```python
df.replace([1,2,3,4,5,6], 6)  # replace different value as 1 common value
```

```python
df.replace([1,2, 'comsat', 'hi', 'Offline'], [3,'ok', 'Malakand', 'bye', 'online'])  # replace multiple values
```

```python
df.replace({'profit': 'hi', 'Sales Channel': 'Online'}, 'none') # using dictionry
```

```python
df.replace('[A-Za-z]', 1 , regex = True)  # replace All string As int e.g here we replace all string as 1
```

```python
df.replace({'Country':'[A-Za-z]'}, 1 , regex = True)  # replace All string of columns (using dict ) as int e.g here we replace all string as 1
```

```python
df.replace('no', method = 'bfill', limit = 2) # using limit
```

```python
df.replace('hi', 'ok', inplace = True)  # replace in orginal datasets  not making new datasets
df
```
