## Merging Function

```python
df1 = pd.DataFrame({'ID': [1,2,3,4],            # creating Data Frame
                 'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID': [1,2,3,4],
                   'Name': ['A', 'B', 'C', 'D']})
df2
```

```python
df3 = pd.DataFrame({'ID': [1,2,3,5],         
                   'Roll#': [44,45,46,47]})
df3
```

```python
pd.merge(df2, df1 )          # merging Data Frame
```

```python
pd.merge(df2, df3 )         # ID 4th row not printed
```

```python
  pd.merge(df2, df3, how = 'right' )    # print the left whole row  printed
```

```python
  pd.merge(df2, df3, how = 'right' )         # print right whole row
```

```python
  pd.merge(df2, df3, how = 'outer' )         # print both right and left whole row
```

```python
  pd.merge(df2, df3, how = 'outer' , indicator = True)         #Just Highlight values which is available in which columns
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4],         
                 'Class': [5,6,7,8]})

df2 = pd.DataFrame({'ID': [5,6,7,8],
                   'Name': ['A', 'B', 'C', 'D']})
```

## Joind DataFrames

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B': [5,6,7,8]})
df2 = pd.DataFrame({'C': [11,2,22,23],
                   'D': [13,12,44,56]})
df1.join(df2)   # joing data frames
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B': [5,6,7,8]}, index = [5,6,7,8])     # becuse of differenet index of both DataFrams the Data aree not joinig correctly
df2 = pd.DataFrame({'C': [11,2,22,23],
                   'D': [13,12,44,56]})  # make sure index will be same of both dataFrame 
df1.join(df2)   # joing data frames
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4,6],
                   'B': [5,6,7,8, 5]})
df2 = pd.DataFrame({'C': [11,2,22,23],
                   'D': [13,12,44,56]})
df1.join(df2)  # both dataframe are not of same length thats why NAN column in last row of output for handling this problem we use how funtions
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4,6],
                   'B': [5,6,7,8, 5]})
df2 = pd.DataFrame({'C': [11,2,22,23],
                   'D': [13,12,44,56]})
df1.join(df2, how = 'right')  # print the right hand side dataFrames number of rows
```

```python
df1.join(df2, how = 'left')  # print the left hand side dataFrames number of rows
```

```python
df1.join(df2, how = 'inner')  # inner same work as LHS 
```

```python
df1.join(df2, how = 'outer')  # inner same work as RHS 
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B': [5,6,7,8]})
df2 = pd.DataFrame({'A': [11,2,22,23],
                   'D': [13,12,44,56]})
df1.join(df2, lsuffix = '1')   # if dataframe have same columns names then we use lsuffix or rsuffix for chamging columns names
```

```python
df1 = pd.DataFrame({'A': [1,2,3,4],
                   'B': [5,6,7,8]})
df2 = pd.DataFrame({'A': [11,2,22,23],
                   'D': [13,12,44,56]})
df1.join(df2, rsuffix = '1')   # same as above examaple
```
